```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "iframe"    # alternative, safe for large plots
from pathlib import Path
import glob
```

```python
def user_prot_inference_merged_per_sample(acc):
    print(f"Processing bakta data for {acc}")
    inference=pd.read_csv(
        f"/hpf/largeprojects/ccmbio/ajain/isaac_chantel_project/pipelines/pipeline_runs/{acc}/annotated/{acc}_bakta_PI_30/{acc}_PI_30.inference.tsv",
        sep="\t",skiprows=5
        )
    
    print(inference.shape)
        
    all_annots=pd.read_csv(
        f"/hpf/largeprojects/ccmbio/ajain/isaac_chantel_project/pipelines/pipeline_runs/{acc}/annotated/{acc}_bakta_PI_30/{acc}_PI_30.tsv",
        sep="\t",skiprows=5
        )
    print(all_annots.shape)
    
    user_prot_annots=all_annots[all_annots["DbXrefs"].str.contains("UserProtein",na=False)]
    
    print(user_prot_annots.shape)
    
    user_proteins_with_inference=user_prot_annots.merge(inference,on=["#Sequence Id","Type","Start","Stop","Strand"])
    print(user_proteins_with_inference.shape)
    
    if user_prot_annots.shape[0] != user_proteins_with_inference.shape[0]:
        print("Rows are not matching up! Please check")
    
    user_proteins_with_inference['Id'] = pd.to_numeric(user_proteins_with_inference['Id'])
    
    agg_data = (
        user_proteins_with_inference
        .groupby(['Accession', 'Product'], dropna=False)
        .agg(
            count=('Id', 'size'),
            min_identity=('Id', 'min'),
            max_identity=('Id', 'max'),
            n_ge_90=('Id', lambda x: (x >= 0.90).sum()),
            n_lt_90=('Id', lambda x: (x < 0.90).sum())
        )
        .reset_index()
    )
    
    
    # Add gene names
    user_gene_annot=pd.read_excel("/hpf/largeprojects/ccmbio/ajain/isaac_chantel_project/pipelines/data/pseudomonas_annot_protein_list_curated/User_provided_annotations.xlsx").drop(["Accession_2","Product"],axis=1)

    agg_data.merge(user_gene_annot,on=["Accession"])
    
    agg_data["Sample"]=acc
    
    agg_data.to_csv(f"/hpf/largeprojects/ccmbio/ajain/isaac_chantel_project/pipelines/pipeline_runs/bakta_PI_30_agg_results/{acc}.csv",index=False)
    
    
    print("Done!")
```

```python
# generate user_protein_inference merged files for all SRA runs
acc_list=pd.read_csv("/hpf/largeprojects/ccmbio/ajain/isaac_chantel_project/pipelines/pipeline_runs/acc_list/SraAccList.csv")

for acc in acc_list["acc"].tolist():
    user_prot_inference_merged_per_sample(acc)
```

## Generate a heatmap

```python
# Path to the folder with CSV files
folder = Path("/hpf/largeprojects/ccmbio/ajain/isaac_chantel_project/pipelines/pipeline_runs/bakta_PI_30_agg_results/")

# Get all CSV files in the folder
csv_files = list(folder.glob("*.csv"))

# Read each CSV and concatenate them row-wise
combined_df = pd.concat([pd.read_csv(f) for f in csv_files], ignore_index=True)

# Save the combined dataframe
combined_df.to_csv("/hpf/largeprojects/ccmbio/ajain/isaac_chantel_project/pipelines/pipeline_runs/bakta_PI_30_agg_results/combined_for_all_samples.csv", index=False)
```

```python
# Heatmap for Occurrences  90% Identity

# Pivot tables
heat_ge90 = combined_df.pivot_table(index='Accession', columns='Sample', values='n_ge_90', fill_value=0)

fig = px.imshow(
    heat_ge90,
    labels=dict(x="Sample", y="Accession", color="n_ge_90"),
    aspect="auto",
    color_continuous_scale="Viridis"
)

fig.update_layout(
    title="Occurrences  90% Identity",
    height=1600,
    width=1600
)

# Hide sample names on the X axis
fig.update_xaxes(showticklabels=False)

fig.update_yaxes(
    tickmode='array',
    tickvals=list(range(len(heat_ge90.index))),
    ticktext=heat_ge90.index.tolist()
)

fig.show()
```

```python
# Heatmap for Occurrences < 90% Identity

heat_lt90 = combined_df.pivot_table(index='Accession', columns='Sample', values='n_lt_90', fill_value=0)

fig = px.imshow(
    heat_lt90,
    labels=dict(x="Sample", y="Accession", color="n_lt_90"),
    aspect="auto",
    color_continuous_scale="Viridis"
)

fig.update_layout(
    title="Occurrences < 90% Identity",
    height=1600,
    width=1600
)

# Hide sample names on the X axis
fig.update_xaxes(showticklabels=False)

fig.update_yaxes(
    tickmode='array',
    tickvals=list(range(len(heat_lt90.index))),
    ticktext=heat_lt90.index.tolist()
)


fig.show()
```
